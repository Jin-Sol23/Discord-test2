# JSON

```python
# json 문자열을 json.loads() 함수를 사용하여 파이썬 객체로 변환
import json

d4_str = """
[
  {"name": "혼자 공부하는 데이터 분석", "author": "박해선", "year": 2022},
  {"name": "혼자 공부하는 머신러닝+딥러닝", "author": "박해선", "year": 2020}
]
"""
d4 = json.loads(d4_str)
print(d4[0]['name'])
```

```python
# json 문자열을 read_json() 함수를 사용하여 판다스 데이터프레임으로 변환
import pandas as pd
pd.read_json(d4_str)
```

# 판다스 데이터프레임 - 열 삭제

```python
# 남산도서관 공공 데이터를 다운로드
import gdown
gdown.download('https://bit.ly/3RhoNho', 'ns_202104.csv', quiet=False)
```

```python
# 판다스 데이터프레임으로 읽은 후, 처음부터 5개 행을 출력
import pandas as pd
ns_df = pd.read_csv('ns_202104.csv', low_memory=False)
ns_df.head()
```

```python
# loc 메서드에 슬라이싱을 사용하여 Unnamed:13 열을 제외한 새로운 데이터프레임 생성
ns_book = ns_df.loc[:, '번호':'등록일자']
ns_book.head()
```

```python
# 컬럼의 속성 확인 = index 클래스 객체
print(ns_df.columns)
```

```python
# 리스트와 같이 숫자 인덱스로 참조 가능
print(ns_df.columns[0])
```

```python
# 원소별 비교 : index 클래스, 판다스 배열 성격의 객체 등을 자동으로 배열에 있는 모든 원소와 하나씩 비교
ns_df.columns != 'Unnamed: 13'
# 반환된 결과는 넘파이 배열이다.
```

```python
# 원소별 비교를 통해 불리언 배열을 사용할 수 있음 (슬라이싱은 중간에 있거나 특정 열을 제외하기 어려움)
selected_columns = ns_df.columns != 'Unnamed: 13'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
# drop() 메서드를 사용하여 Unnamed: 13 열 삭제
# axis 매개변수 지정 : 0(기본값, 행 삭제), 1(열 삭제)
ns_book = ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()
```

```python
# 리스트 형식을 사용하면 여러 열 삭제 가능
ns_book = ns_df.drop(['부가기호', 'Unnamed: 13'], axis=1)
ns_book.head()
```

```python
# inplace 매개변수를 True로 지정하면, 현재 선택한 데이터프레임을 바로 수정 가능함
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

```python
# dropna() 메서드 : 행, 열에 NaN(비어있는 값)이 하나 이상 포함되면 삭제함
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
# dropna() 메서드에서 how 매개변수를 'all'로 지정하여 모든 값이 NaN인 열 삭제
ns_book = ns_df.dropna(axis=1, how='all')
ns_book.head()
```
